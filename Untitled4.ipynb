```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer


# create a function to compare different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
  model = RandomForestRegressor(n_estimators=10, random_state=0)
  model.fit(X_train, y_train)
  pred = model.predict(X_valid)
  return mean_absolute_error(y_valid, pred)


#load data file
df = pd.read_csv("/content/aviation_sales_data.csv", skipinitialspace=True)



#retrive all info
print(df.head())
print(df.describe())
print(df.info())
print(df.shape)



#calculate the amount of missing values per category and % overall
missing_values = df.isnull().sum()
percentage_of_category =  round((missing_values / len(df)),2) * 100
print(f"The total amount of missing values: \n{missing_values}")
print(f"The percentage per category is: \n{percentage_of_category}")



#remove duplicated rows
df = df.drop_duplicates()
print(f"Duplicated data: \n {len(df)}")


#remove NAN values in revenue booking
df = df.dropna(subset=['revenue_eur'])


#fix date format first change type from object to str
print(df['booking_date'])
df['booking_date'] = df['booking_date'].astype('str')
df['booking_date'] = pd.to_datetime(df['booking_date'], format='mixed').dt.strftime('%m')
print(df['booking_date'])


#check for inconsistency to clean the data further eg. spaces and use lowercase
for col in df.select_dtypes(include='object').columns:
  print(col, df[col].unique())


#make all data lowercase:
for col in df.select_dtypes(include='object').columns:
  df[col] = df[col].str.lower()
  print(df)
print(df.columns)



#Features
features = ['booking_date', 'airline', 'route', 'cabin_class', 'market',
       'sales_channel', 'trip_type', 'tickets_sold']


#Target
y = df['revenue_eur']
X = df[features]
X.head()

# Split data into train (x) and validation (y)
train_X, val_X, train_y, val_y = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=1)



# numerical columns and categorial
numerical_cols = [col for col in train_X.columns if train_X[col].dtype in ['int64', 'float64']]
categorial_cols = [col for col in train_X.columns if train_X[col].dtype == 'object']



# preprocessing for numerical data
numerical_data = SimpleImputer(strategy="mean")


# preprocessing categorial data
categorial_transformed = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                                         ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# combine the numerical and categorial data
preprocessor = ColumnTransformer(transformers=[('num', numerical_data, numerical_cols),
                                               ('cat', categorial_transformed, categorial_cols)])

# create the model
# model = RandomForestRegressor(n_estimators=100, random_state=0)
model = XGBRegressor(n_estimators=200, random_state=40, learning_rate=0.03, max_depth=3 )

# final pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])


# fit on trainingsdata
pipeline.fit(train_X, train_y)

# predict on validation data
preds = pipeline.predict(val_X)

# calculate MAE
mae = mean_absolute_error(val_y, preds)
print(f"The mean Absolute error is: {mae}")

print(y.mean())
print(y.describe())
cabin_class = df.groupby('cabin_class')['revenue_eur'].mean()
print(cabin_class)


comparison = pd.DataFrame({'Actual value:': val_y,
                           'Predicted Value:': preds})
print(comparison.head(10))

```
